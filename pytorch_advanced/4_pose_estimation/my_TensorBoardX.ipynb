## graphファイルの作成

```python
from utils.openpose_net import OpenPoseNet
# モデルの用意
net = OpenPoseNet()
net.train()
```

```python
# graphファイルを保存するためのクラスwriterを用意
# 1. tensorboardXの保存クラスを呼び出す
from tensorboardX import SummaryWriter

# 2. ディレクトリ tbXに保存させるwriterを用意
# tbXがなければ自動で生成される
writer = SummaryWriter("./tbX/")

# 3. ネットワークに流し込むダミーデータを作成
batch_size = 2
dummy_img = torch.rand(batch_size, 3, 368, 368)

# 4. OpenPoseのインスタンスnetに対してdummy_imgを流したときのgraphをwriterに保存
writer.add_graph(net, (dummy_img, ))
writer.close()

# 5. ターミナルでtbXがあるディレクトリまで移動して以下のコマンドを実行
# tensorboard --logdir="./tbX/"
# http://localhost:6006

# localではRuntime errorになる　お手本も同じエラー　　awsの環境でうまくいくか試してみる
```
